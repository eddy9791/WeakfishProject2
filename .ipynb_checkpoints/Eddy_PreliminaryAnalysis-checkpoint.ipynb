```python
%matplotlib notebook
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
filename = 'LipidData.csv' 
data = pd.read_csv(filename,header=0,na_values=' ')
data.head()
```

```python
plt.figure()
plt.plot(data['SL'],data['DRYWT'],'c.')
plt.title('Weakfish Length and Weight curve')
plt.ylabel('Dry Weight (mg)')
plt.xlabel('Standard Length (mm)');
```

```python
plt.figure()
plt.plot(data['SL'],data['TAGPERTOTLIP'],'.')
plt.title('Percent of Triacylglyceride  by standard length')
plt.xlabel('Standard Length (mm)')
plt.ylabel('TAG (percent of total lipid content)');
```

```python
## index based on location

bm = data['LOC$'] == 'BM'
ac = data['LOC$'] == 'AC'
lb = data['LOC$'] == 'LB'
lbc = data['LOC$'] == 'LBC'
mb = data['LOC$'] == 'MB'
mhc = data['LOC$'] == 'MHC'
ub = data['LOC$'] == 'UB'

## plot the length vs tag curve based on location
plt.figure()
plt.plot(data['SL'][bm],data['TAGPERTOTLIP'][bm],'r.',label = 'BM')
plt.plot(data['SL'][ac],data['TAGPERTOTLIP'][ac],'g.', label = 'AC')
plt.plot(data['SL'][lb],data['TAGPERTOTLIP'][lb],'b.', label = 'LB')
plt.plot(data['SL'][lbc],data['TAGPERTOTLIP'][lbc], 'c.', label = 'LBC')
plt.plot(data['SL'][mb],data['TAGPERTOTLIP'][mb],'y.', label = 'MB')
plt.plot(data['SL'][mhc],data['TAGPERTOTLIP'][mhc],'m.', label = 'MHC')
plt.plot(data['SL'][ub],data['TAGPERTOTLIP'][ub],'k.', label = 'UB')
plt.legend(title = "Location")
plt.xlabel('Standard Length (mm)')
plt.ylabel('TAG (percent of total lipid concentration)');
```

```python
plt.figure()
plt.plot(data['C'][bm],data['N'][bm],'r.',label = 'BM')
plt.plot(data['C'][ac],data['N'][ac],'g.', label = 'AC')
plt.plot(data['C'][lb],data['N'][lb],'b.', label = 'LB')
plt.plot(data['C'][lbc],data['N'][lbc], 'c.', label = 'LBC')
plt.plot(data['C'][mb],data['N'][mb],'y.', label = 'MB')
plt.plot(data['C'][mhc],data['N'][mhc],'m.', label = 'MHC')
plt.plot(data['C'][ub],data['N'][ub],'k.', label = 'UB')
plt.legend(title = "Location")
plt.xlabel('$\delta^{13}$C')
plt.ylabel('$\delta^{15}$N')
plt.title( 'Isotope Plot');
```
